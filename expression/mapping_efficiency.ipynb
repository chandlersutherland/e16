```python
import os
import pandas as pd 
import glob
```

```python
final_output=glob.glob('/global/scratch/users/chandlersutherland/e16/*/rna_tip/STAR/*_Log.final.out')
```

```python
accession=final_output[0].split('/')[-1].split('_')[0]
sample=final_output[0].split('/')[6]
```

```python
#initialize final df 
mapping_stats=pd.DataFrame(columns=['sample', 'accession', 'n_input_reads', 'avg_input_read_length', 'n_unique_mapping', 'percent_unique_mapping', 'percent_multimapping'])

for i in range(0, len(final_output)):
    #save accession and sample information 
    accession=final_output[i].split('/')[-1].split('_')[0]
    sample=final_output[i].split('/')[6]
    
    #read file 
    with open(final_output[i]) as f:
        lines=f.readlines()
    
    #pull out lines of interest 
    input_reads = [i for i in lines if 'input read' in i]
    unique_mapper = [i for i in lines if 'Uniquely mapped reads' in i]
    multi_mapper = [i for i in lines if '% of reads mapped to multiple loc' in i]
    of_interest = input_reads + unique_mapper + multi_mapper

    #get values of each line  
    values=[sample, accession]
    for j in range(0, len(of_interest)):
        val=of_interest[j].replace('\t', '').replace('\n','').replace('%', '').split('|')[1]
        values.append(val)
    
    mapping_stats.loc[i] = values 
    
mapping_stats
```

```python
#write out df to tsv 
mapping_stats.to_csv('/global/scratch/users/chandlersutherland/e16/cs_reports/rna_tip_mapping_stats.tsv', sep='\t')
```
